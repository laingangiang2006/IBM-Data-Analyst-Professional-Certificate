### Task 1: Visualizing Data Distributions


##### 1. Histogram of `ConvertedCompYearly`


Visualize the distribution of yearly compensation (`ConvertedCompYearly`) using a histogram.



```python
## Write your code here
comp = df['ConvertedCompYearly'].dropna()

plt.figure(figsize=(10,6))
plt.hist(comp, bins=30)

plt.title("Distribution of Yearly Compensation")
plt.xlabel("Converted Compensation Yearly")
plt.ylabel("Frequency")

plt.show()
```

##### 2. Box Plot of `Age`


Since `Age` is categorical in the dataset, convert it to numerical values for a box plot.



```python
## Write your code here
df['Age_numeric'] = pd.to_numeric(df['Age'], errors='coerce')

age_data = df['Age_numeric'].dropna()

plt.figure(figsize=(8,5))
plt.boxplot(age_data)

plt.title("Box Plot of Age")
plt.ylabel("Age")

plt.show()
```

### Task 2: Visualizing Relationships in Data


##### 1. Scatter Plot of `Age_numeric` and `ConvertedCompYearly`


Explore the relationship between age and compensation.



```python
## Write your code here
scatter_data = df[['Age_numeric','ConvertedCompYearly']].dropna()

plt.figure(figsize=(10,6))
plt.scatter(scatter_data['Age_numeric'], scatter_data['ConvertedCompYearly'])

plt.title("Age vs Yearly Compensation")
plt.xlabel("Age")
plt.ylabel("Converted Compensation Yearly")

plt.show()
```

##### 2. Bubble Plot of `ConvertedCompYearly` and `JobSatPoints_6` with `Age_numeric` as Bubble Size


Explore how compensation and job satisfaction are related, with age as the bubble size.


```python
## Write your code here
bubble_data = df[['ConvertedCompYearly','JobSatPoints_6','Age_numeric']].dropna()

plt.figure(figsize=(10,6))
plt.scatter(
    bubble_data['ConvertedCompYearly'],
    bubble_data['JobSatPoints_6'],
    s=bubble_data['Age_numeric']*2,
    alpha=0.5
)

plt.title("Compensation vs Job Satisfaction")
plt.xlabel("Converted Compensation Yearly")
plt.ylabel("Job Satisfaction Score")

plt.show()
```

### Task 3: Visualizing Composition of Data with Bar Charts


##### 1. Horizontal Bar Chart of `MainBranch` Distribution


Visualize the distribution of respondents’ primary roles to understand their professional focus.



```python
## Write your code here
branch_counts = df['MainBranch'].value_counts()

plt.figure(figsize=(10,6))
branch_counts.plot(kind='barh')

plt.title("Distribution of Main Professional Branch")
plt.xlabel("Count")
plt.ylabel("MainBranch")

plt.show()
```

##### 2. Vertical Bar Chart of Top 5 Programming Languages Respondents Want to Work With


Identify the most desired programming languages based on `LanguageWantToWorkWith`.



```python
## Write your code here
languages = df['LanguageWantToWorkWith'].dropna()

lang_series = languages.str.split(';').explode()

top_languages = lang_series.value_counts().head(5)

plt.figure(figsize=(8,6))
top_languages.plot(kind='bar')

plt.title("Top 5 Programming Languages Developers Want to Work With")
plt.xlabel("Programming Language")
plt.ylabel("Count")

plt.show()
```

##### 3. Stacked Bar Chart of Median `JobSatPoints_6` and `JobSatPoints_7` by Age Group


Compare job satisfaction metrics across different age groups with a stacked bar chart.


```python
## Write your code here
sat_data = df[['Age','JobSatPoints_6','JobSatPoints_7']].dropna()

sat_group = sat_data.groupby('Age')[['JobSatPoints_6','JobSatPoints_7']].median()

sat_group.plot(kind='bar', stacked=True, figsize=(12,6))

plt.title("Job Satisfaction by Age Group")
plt.xlabel("Age")
plt.ylabel("Median Job Satisfaction")

plt.show()
```

##### 4. Bar Chart of Database Popularity (`DatabaseHaveWorkedWith`)


Identify the most commonly used databases among respondents by visualizing `DatabaseHaveWorkedWith`.



```python
## Write your code here
db_data = df['DatabaseHaveWorkedWith'].dropna()

db_series = db_data.str.split(';').explode()

top_db = db_series.value_counts().head(10)

plt.figure(figsize=(10,6))
top_db.plot(kind='bar')

plt.title("Most Popular Databases")
plt.xlabel("Database")
plt.ylabel("Count")

plt.show()
```

### Task 4: Visualizing Comparison of Data with Bar Charts


##### 1. Grouped Bar Chart of Median `ConvertedCompYearly` for Different Age Groups


Compare median compensation across multiple age groups with a grouped bar chart.



```python
## Write your code here
comp_age = df[['Age','ConvertedCompYearly']].dropna()

median_comp = comp_age.groupby('Age')['ConvertedCompYearly'].median()

plt.figure(figsize=(12,6))
median_comp.plot(kind='bar')

plt.title("Median Compensation by Age Group")
plt.xlabel("Age")
plt.ylabel("Median Compensation")

plt.show()
```

##### 2. Bar Chart of Respondent Count by Country


Show the distribution of respondents by country to see which regions are most represented.



```python
## Write your code here
country_counts = df['Country'].value_counts().head(10)

plt.figure(figsize=(10,6))
country_counts.plot(kind='bar')

plt.title("Top Countries by Respondent Count")
plt.xlabel("Country")
plt.ylabel("Number of Respondents")

plt.show()
```
